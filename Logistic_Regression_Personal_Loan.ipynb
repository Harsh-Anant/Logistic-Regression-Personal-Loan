```python
import pandas as pd
```

```python
dataframe=pd.read_excel('Downloads/Bank_Personal_Loan_Modelling.xlsx',sheet_name=1)
```

```python
dataframe
```

```python
Y=dataframe['Personal Loan']
```

```python
X=dataframe
```

```python
X.drop('Personal Loan', axis=1, inplace=True)
```

```python
X
```

```python
import statsmodels.api as sm
```

```python
X1=sm.add_constant(X)
```

```python
logistic=sm.Logit(Y,X1)
```

```python
result=logistic.fit()
```

```python
result.summary()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X.drop('ID', axis=1, inplace=True)
X.drop('Age', axis=1, inplace=True)
X.drop('Experience', axis=1, inplace=True)
X.drop('ZIP Code', axis=1, inplace=True)
X.drop('Mortgage', axis=1, inplace=True)
'''all the above dropped attributes having their P>0.05, so for this reason they have been dropped beacuse so don.t have any significant
effect on Personal Loan.'''
```

```python
X
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
Xtrain1=sm.add_constant(Xtrain)
```

```python
logistic=sm.Logit(Ytrain,Xtrain1)
```

```python
result=logistic.fit()
```

```python
result.summary()
```

```python
Xtest1=sm.add_constant(Xtest)
yhat=result.predict(Xtest1)
```

```python
prediction=list(map(round,yhat))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
res=confusion_matrix(Ytest,prediction)
```

```python
res
```

```python
acc=accuracy_score(Ytest,prediction)
```

```python
acc*100
```

```python
cla=classification_report(Ytest,prediction)
```

```python
print(cla)
```
